```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import cv2
img =cv2.imread('download.jpg')
imgplot = plt.imshow(img)
plt.show()
print('The shape of the image is {0}'.format(img.shape))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#Translate the image to (X,Y)


```python
def translate(img,x,y):
  img.astype(np.int32)
  h,w=img.shape
  filter=np.array([[1,0,x],[0,1,y]])
  output = np.zeros(img.shape , dtype= 'u1')
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i,1])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
     
      if 0 <= x0 < w and 0 <= y0 < h:
        output[y0,x0] = img[i,j]
  return output

figure, axis = plt.subplots(1,2)
axis[0].imshow(gray)
axis[0].set_title("orignal image")
trans_img=translate(gray,10,40)
axis[1].imshow(trans_img)
axis[1].set_title("translated image")
plt.show()
```

#Scale about the origin

```python
def scale_origin(img,W,H):
  img.astype(np.int32)
  h,w=img.shape
  filter=np.array([[W,0],[0,H]])
  output = img
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      if 0 <= x0 < w and 0 <= y0 < h:
        output[y0,x0] = img[i,j]
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return blur


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
figure, axis = plt.subplots(1,2)
axis[0].imshow(gray[:])
axis[0].set_title("orignal image")
op_scaled=scale_origin(gray,1.2,0.8)
axis[1].imshow(op_scaled)
axis[1].set_title("scaled image")
plt.show()
```

#Rotate about the origin

```python
def rotate_origin(img,theta):
  img.astype(np.int32)
  h,w=img.shape
  cos=np.cos(np.pi/180 * theta)
  sin=np.sin(np.pi/180 * theta)
  filter=np.array([[cos,sin],[-sin,cos]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      if 0 <= x0 < w and 0 <= y0 < h:
        output[y0,x0] = img[i,j]
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
figure, axis = plt.subplots(1,2)
axis[0].imshow(gray[:])
axis[0].set_title("orignal image")
# Angle in degrees
op_scaled=rotate_origin(gray,5)
axis[1].imshow(op_scaled)
axis[1].set_title("rotated image")
plt.show()
```

# Sheer in x and y direction 

```python
def sheer_x(img,theta):
  img.astype(np.int32)
  h,w=img.shape
  tan=np.tan(np.pi/180 * theta)
  filter=np.array([[1,tan],[0,1]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      if 0 <= x0 < w and 0 <= y0 < h:
        output[y0,x0] = img[i,j]
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output


def sheer_y(img,theta):
  img.astype(np.int32)
  h,w=img.shape
  tan=np.tan(np.pi/180 * theta)
  filter=np.array([[1,0],[tan,1]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      if 0 <= x0 < w and 0 <= y0 < h:
        output[y0,x0] = img[i,j]
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
figure, axis = plt.subplots(1,3)
axis[0].imshow(gray[:])
axis[0].set_title("orignal image")
# Angle in degrees

sh_x=sheer_x(gray,5)
axis[1].imshow(sh_x)
axis[1].set_title("sheer in x")

sh_y=sheer_y(gray,9)
axis[2].imshow(sh_y)
axis[2].set_title("sheer in y")

plt.show()
```

# Reflection

```python
def reflect_origin(img):
  img.astype(np.int32)
  h,w=img.shape
  filter=np.array([[-1,0],[0,-1]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      output[y0,x0] = img[i,j]
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output

def reflect_x(img):
  img.astype(np.int32)
  h,w=img.shape
  filter=np.array([[1,0],[0,-1]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      output[y0+img.shape[1]-1,x0] = img[i,j]
      #print(x0,y0)
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output

def reflect_y(img):
  img.astype(np.int32)
  h,w=img.shape
  filter=np.array([[-1,0],[0,1]])
  output = np.zeros(img.shape)
  for i in range(h):
    for j in range(w):
      xy=np.array([j,i])
      n_xy=np.dot(filter,xy)
      x0=int(n_xy[0])
      y0=int(n_xy[1])
      output[y0,x0+img.shape[0]-1] = img[i,j]
      
  scaled=output[0:y0+1,0:x0+1]
  blur = cv2.GaussianBlur(scaled,(3,3),0)     
  return output


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
figure, axis = plt.subplots(2,2)
axis[0][0].imshow(gray)
axis[0][0].set_title("orignal image")
# Angle in degrees



ref_o=reflect_origin(gray)
axis[0][1].imshow(ref_o)
axis[0][1].set_title("reflection about origin")


ref_x=reflect_x(gray)
axis[1][0].imshow(ref_x)
axis[1][0].set_title("reflection about x-axis")



ref_y=reflect_y(gray)
axis[1][1].imshow(ref_y)
axis[1][1].set_title("reflection about y-axis")
plt.show()
```
